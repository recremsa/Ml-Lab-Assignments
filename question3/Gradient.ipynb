Perform linear regression with one feature using gradient descent (without using library function)
to predict the salary of an employee based on the feature Years Experience. Use hyper-parameter
tuning for the best result. Plot the hypothesis function and the data points after each epoch. Evaluate
the accuracy of the models by computing relevant performance metrics, including the R² value.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('salary_data.csv')

X = data['YearsExperience'].values
y = data['Salary'].values

m = len(y)

print(f"Loaded {m} data points.")
```

```python
theta0 = 0
theta1 = 0

lr = 0.01
epochs = 1
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(X, y, color='blue', label='Actual Data')
line, = ax.plot(X, theta0 + theta1 * X, color='red', label='Hypothesis')

plt.title(f'Epoch 0')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()

for epoch in range(epochs):
    y_pred = theta0 + theta1 * X

    d0 = (-2/m) * sum(y - y_pred)
    d1 = (-2/m) * sum(X * (y - y_pred))

    theta0 -= lr * d0
    theta1 -= lr * d1

    line.set_ydata(theta0 + theta1 * X)
    ax.set_title(f'Epoch {epoch+1}')
    fig.canvas.draw()
    plt.pause(0.001)

print("Training complete.")
print(f"Final Parameters: θ0={theta0:.2f}, θ1={theta1:.2f}")
```

```python
y_final = theta0 + theta1 * X

ss_total = sum((y - np.mean(y))**2)
ss_residual = sum((y - y_final)**2)
r2 = 1 - (ss_residual / ss_total)

mae = (1/m) * sum(np.abs(y - y_final))
rmse = np.sqrt((1/m) * sum((y - y_final)**2))

print(f"--- Final Model Performance ---")
print(f"R-squared (R²): {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
