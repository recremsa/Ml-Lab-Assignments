Design and train a network of perceptrons that computes the functionality of XOR.


```python
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])
```

```python
model = MLPClassifier(hidden_layer_sizes=(2,), max_iter=2000, random_state=42).fit(X, y)
```

```python
print("Predictions:", model.predict(X))
print("Accuracy:", model.score(X, y))
```
